# 01 · What Is a Token?

Companion notebook for **[What Is a Token?](https://jayantkhd.vercel.app/blog/transformers-from-scratch/what-is-a-token)**.

A token is a chunk of text from a fixed vocabulary. Common words are usually a single token, while rare words get split into several. The model never sees letters — it sees a sequence of integer token IDs.

```python
# Rough rule of thumb for English: ~4 characters per token
def estimate_tokens(text: str) -> int:
    return max(1, round(len(text) / 4))

sample = "Tokenization is unfamiliar."
print(sample, '->', estimate_tokens(sample), 'tokens (estimate)')
print(1000, 'words ->', round(1000 / 0.75), 'tokens (estimate)')
```

Now the exact counts with a real tokenizer. This needs `tiktoken` (`pip install tiktoken`); the cell skips gracefully if it isn't installed.

```python
try:
    import tiktoken
    enc = tiktoken.get_encoding('cl100k_base')
    ids = enc.encode(sample)
    print('token ids:', ids)
    print('pieces:  ', [enc.decode([i]) for i in ids])
    print('count:   ', len(ids), 'tokens (exact)')
except ImportError:
    print('tiktoken not installed - run: pip install tiktoken')
```

Notice how `"unfamiliar"` shatters into several pieces while `"is"` stays whole — that asymmetry is exactly why token counts never match word counts.
